# question no:7

```python
#7)
#analysis of variance between etest p mark and mba p mark 
import scipy.stats as stats
stats.f_oneway(dataset['etest_p'],dataset['mba_p'])
```

```python
# it is a one way classification 
```

```python
stats.f_oneway(dataset['etest_p'],dataset['mba_p']).pvalue
```

```python
#hypothesis testing 
print("null hypothesis:there is no significant similarities  between etest_p mark and mba_p mark")
print("alternate hypothesis:there is significant similarities between etest_p mark and mba_p mark")
if(stats.f_oneway(dataset['etest_p'],dataset['mba_p']).pvalue<0.05):
    print("accept null hypothesis")
    
else:
    
    print("accept alternate hypothesis")
    
```

# question no:8

```python
# 8)test the similarities between the degree t(sci&tech) and specializtion(mkt&hr) with respect to salary
```

```python
#it is independant sample (unpaired t test) because of two difrenrt column and same conditions
```

```python
from scipy.stats import ttest_ind
dataset=dataset.dropna()
degree=dataset[dataset['degree_t']=='Sci&Tech']['salary']
specialaization=dataset[dataset['specialisation']=='Mkt&HR']['salary']
ttest_ind(degree,specialaization)
```

```python
ttest_ind(degree,specialaization).pvalue
```

```python
print("null hypothesis:there is no significant similarities  between degree t(sci&tech)and specializtion(mkt&hr)with salary")
print("alternate hypothesis:there is significant similarities between degree t(sci&tech) and specializtion(mkt&hr) with  salary")
     
if(ttest_ind(degree,specialaization).pvalue<0.05):
    print("accept null hypothesis")
   
else:
   
    print("accept alternate hypothesis")
    
```

```python
#to convert above normal distribution to standard normal distribution
#this is a coding for std normal distribution plot
def stdNBgraph(dataset):
    import seaborn as sns
    mean=dataset.mean()
    std=dataset.std()
    values=[i for i in dataset]
    z_score=[((j-mean)/std)for j in values]
    sns.distplot(z_score)/len(z_score)
```

```python
stdNBgraph(dataset["salary"])
```

```python
#above graph shown standard normal distribution of salary column without change the normal distribution plot information
#only changed the x axis values but not changed information that is called some defined range of x axis
```

# question no:10

```python
#what is the probability density function of the salary range from 700000 to 900000
```

```python
def get_pdf_probability(dataset,startrange,endrange): 
    from matplotlib import pyplot
    from scipy.stats import norm
    import seaborn as sns
    ax=sns.distplot(dataset,kde=True,kde_kws={'color':'blue'},color='Green')
    pyplot.axvline(startrange,color='Red')
    pyplot.axvline(endrange,color='Red')
    sample=dataset
    sample_mean=sample.mean()
    sample_std=sample.std()
    print('mean=%.3f,Standard Deviation=%.3f'%(sample_mean,sample_std))
    dist=norm(sample_mean,sample_std)
    values=[value for value in range(startrange,endrange)]
    probabilities=[dist.pdf(value) for value in values]
    prob=sum(probabilities)
    print("the area beatween range({},{}:{}".format(startrange,endrange,sum(probabilities)))
    return prob  
```

```python
get_pdf_probability(dataset["salary"],700000,900000)
```
